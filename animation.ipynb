```python
%matplotlib widget

from pylab import *
from time import sleep
import ipywidgets as widgets
from IPython.display import display

# Initialisation des tableaux x et y
t, u = [], []

# Configuration de la figure
fig, ax = subplots()
line, = ax.plot([], [], 'r-')  # Trace initiale

# DÃ©finir les limites initiales des axes
ax.set_xlim(0, 10)
ax.set_ylim(-1, 1)

def update_graph(t, u):
    line.set_data(t, u)
    ax.relim()
    ax.autoscale_view()
    fig.canvas.draw()
    fig.canvas.flush_events()

def animate_data():
    for i in range(100):
        new_t = i * 0.1
        new_u = sin(new_t)
        
        t.append(new_t)
        u.append(new_u)
        
        # Mettre Ã  jour le graphique
        update_graph(t, u)
        
        # Pause pour voir l'animation en temps rÃ©el
        sleep(0.1)

# Bouton pour dÃ©marrer les mesures
start_button = widgets.Button(description="DÃ©marrer")
def on_button_click(b):
    animate_data()
start_button.on_click(on_button_click)

# Afficher le bouton
display(start_button)
```
